# gmKriger: A Kriging-based ground motion intensity measure calculator
---
## Ground motion IMs for the 1989 M6.7 Loma Prieta Earthquake

### Imports

```python
import pandas as pd
import gmKriger
import geostats
import time
```

### Select backend

```python
backend = 'Cython'
```

### Enter sites' ID, latitude, longitude, and site $V_{S30}$

```python
site = ['Alameda Naval Air Station', 'Treasure Island', 'Alameda Bay Farm Island', 'Farris Farm', 'POO7']
lat  = [37.785748, 37.8261394, 37.73380567, 36.91026828,37.805242]
lon  = [-122.309346, -122.3712351, -122.250101,-121.7437891,-122.339702]
Vs30 = [186.2, 181.1, 230.7, 209.5,223]
```

### Select an earthquake and ground motion intensity measures (GMIMs)

```python
earthquake = '1989 M6.9 Loma Prieta'
gmims = ['PGA','PGV','Ia','CAV']
```

### Select the correlation model type

***realizations***: Select this to use 1000 Bayesian spatial correlation models.<br>
***MAP***: Select this to use the ***M***aximum ***A*** ***P***osteriori spatial correlation model.<br>

```python
model = 'realizations'
```

### Compute ground motions

```python
start = time.time()
```

```python
dfs = []
for gmim in gmims:
    dfs.append(gmKriger.get_Kgmim(site,lat,lon,Vs30,earthquake,model,gmim,backend=backend))
```

```python
end = time.time()
```

### Results

```python
results = pd.merge(dfs[0], dfs[1], on=['Site','Lat (deg)','Lon (deg)'], how='outer')
for df in dfs[2:]:
    results = pd.merge(results, df, on=['Site','Lat (deg)','Lon (deg)'], how='outer')
results.round(3).T
```

```python
print('The calculations using the {} backend took {:.1f} seconds'.format(backend,end-start))
```
